This cell imports necessary libraries (`pandas`, `matplotlib.pyplot`, `seaborn`, `plotly.express`), loads the `amazon_sales_dataset.csv` file into a pandas DataFrame named `df`, and then displays the first 5 rows of the DataFrame to give an initial overview of the data.

```python
# Analyze customer distribution by country
country_counts = df['country'].value_counts().reset_index()
country_counts.columns = ['country', 'customer_count']

# Display the top countries by customer count
display(country_counts.head())

# Plot Customer Distribution by Country using Plotly
fig = px.bar(
    country_counts.nlargest(10, 'customer_count'), # Display top 10 countries
    x='customer_count',
    y='country',
    orientation='h',
    title='Customer Distribution by Country',
    labels={'customer_count': 'Number of Customers', 'country': 'Country'}
)
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

This cell displays a concise summary of the DataFrame `df` using `df.info()`. This output is crucial for understanding the data types of each column, identifying the number of non-null values, and checking the memory usage, which helps in initial data quality assessment.

```python
# Reset the index to ensure 'order_date' is a column if it was previously set as index
df.reset_index(inplace=True, drop=False)

# Convert 'order_date' to datetime objects
df['order_date'] = pd.to_datetime(df['order_date'])

# Set 'order_date' as the index for time-series operations
df.set_index('order_date', inplace=True)

# Resample the data to get monthly total sales (revenue)
monthly_revenue = df['total_sales'].resample('ME').sum().reset_index()
monthly_revenue.columns = ['order_date', 'total_revenue']

# Display the first few rows of the monthly revenue data
display(monthly_revenue.head())

# Plot Revenue over time using Plotly
fig = px.line(
    monthly_revenue,
    x='order_date',
    y='total_revenue',
    title='Revenue Over Time'
)
fig.update_layout(xaxis_title='Date', yaxis_title='Total Revenue')
fig.show()
```

This cell calculates the monthly revenue over time. It first ensures the 'order_date' column is in datetime format and sets it as the DataFrame's index. Then, it resamples the 'total_sales' data monthly ('ME' for month-end) and sums the sales to get `monthly_revenue`. Finally, it displays the first few rows of this data and generates a line plot using Plotly Express to visualize the revenue trend over time.

```python
# Calculate total sales for each product
product_sales = df.groupby('product_name')['total_sales'].sum().nlargest(10).reset_index()
product_sales.columns = ['product_name', 'total_sales']

# Display the top 10 products by sales
display(product_sales)

# Plot Top 10 Products by Sales using Plotly
fig = px.bar(
    product_sales,
    x='total_sales',
    y='product_name',
    orientation='h',
    title='Top 10 Products by Total Sales',
    labels={'total_sales': 'Total Sales', 'product_name': 'Product Name'}
)
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the dataset
df = pd.read_csv('/content/sample_data/amazon_sales_dataset.csv')

# Display the first 5 rows of the DataFrame
display(df.head())
```

This cell analyzes the customer distribution by country. It counts the occurrences of each country in the 'country' column, displays the top countries by customer count, and then generates a horizontal bar chart using Plotly Express to visualize the distribution across the top 10 countries.

```python
# Display basic information about the DataFrame to check data types and missing values
df.info()
```

This cell analyzes the customer distribution by state. It counts the occurrences of each state in the 'state' column, displays the top states by customer count, and then generates a horizontal bar chart using Plotly Express to visualize the distribution across the top 10 states.

```python
# Analyze customer distribution by state
state_counts = df['state'].value_counts().reset_index()
state_counts.columns = ['state', 'customer_count']

# Display the top states by customer count
display(state_counts.head())

# Plot Customer Distribution by State using Plotly
fig = px.bar(
    state_counts.nlargest(10, 'customer_count'), # Display top 10 states
    x='customer_count',
    y='state',
    orientation='h',
    title='Customer Distribution by State',
    labels={'customer_count': 'Number of Customers', 'state': 'State'}
)
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

This cell analyzes total sales for each product category. It groups the DataFrame by 'category', sums the 'total_sales' for each, and then selects the top 10 categories. The results are displayed, and a horizontal bar chart is generated using Plotly Express to visualize these top-performing categories.

```python
# Calculate total sales for each category
category_sales = df.groupby('category')['total_sales'].sum().nlargest(10).reset_index()
category_sales.columns = ['category', 'total_sales']

# Display the top 10 categories by sales
display(category_sales)

# Plot Top 10 Categories by Sales using Plotly
fig = px.bar(
    category_sales,
    x='total_sales',
    y='category',
    orientation='h',
    title='Top 10 Categories by Total Sales',
    labels={'total_sales': 'Total Sales', 'category': 'Product Category'}
)
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

This cell analyzes the total sales for each payment method. It groups the DataFrame by 'payment_method', sums the 'total_sales' for each, and then displays the results. A horizontal bar chart is generated using Plotly Express to visually represent the total sales attributed to each payment method.

```python
# Calculate total shipping costs for each payment method
shipping_cost_by_payment = df.groupby('payment_method')['shipping_cost'].sum().reset_index()
shipping_cost_by_payment.columns = ['payment_method', 'total_shipping_cost']

# Display the shipping costs by payment method
display(shipping_cost_by_payment)

# Plot Shipping Costs by Payment Method using Plotly
fig = px.bar(
    shipping_cost_by_payment,
    x='total_shipping_cost',
    y='payment_method',
    orientation='h',
    title='Total Shipping Costs by Payment Method',
    labels={'total_shipping_cost': 'Total Shipping Cost', 'payment_method': 'Payment Method'}
)
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

This cell analyzes the total shipping costs for each payment method. It groups the DataFrame by 'payment_method', sums the 'shipping_cost' for each, and then displays the results. A horizontal bar chart is generated using Plotly Express to visually compare the shipping costs across different payment methods.

```python
# Calculate total sales for each payment method
payment_method_sales = df.groupby('payment_method')['total_sales'].sum().reset_index()
payment_method_sales.columns = ['payment_method', 'total_sales']

# Display the sales by payment method
display(payment_method_sales)

# Plot Sales by Payment Method using Plotly
fig = px.bar(
    payment_method_sales,
    x='total_sales',
    y='payment_method',
    orientation='h',
    title='Total Sales by Payment Method',
    labels={'total_sales': 'Total Sales', 'payment_method': 'Payment Method'}
)
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

This cell explores the relationship between `discount` and `total_sales` using a scatter plot. It uses Plotly Express to visualize individual data points, where each point represents a transaction's discount percentage and total sales. Hovering over points reveals `product_name` and `category`.

```python
# Explore the relationship between discount and total sales
fig = px.scatter(
    df,
    x='discount',
    y='total_sales',
    title='Relationship Between Discount and Total Sales',
    labels={'discount': 'Discount (%)', 'total_sales': 'Total Sales'},
    hover_data=['product_name', 'category'] # Add additional data to hover tooltip
)
fig.update_layout(xaxis_title='Discount (%)', yaxis_title='Total Sales')
fig.show()
```

This cell calculates the average order value for each payment method. It groups the DataFrame by 'payment_method', calculates the mean of 'total_sales' for each, and then displays the results. A horizontal bar chart is generated using Plotly Express to visually compare the average order values across different payment methods.

```python
# Calculate the average order value for each payment method
avg_order_value_by_payment = df.groupby('payment_method')['total_sales'].mean().reset_index()
avg_order_value_by_payment.columns = ['payment_method', 'average_order_value']

# Display the average order value by payment method
display(avg_order_value_by_payment)

# Plot Average Order Value by Payment Method using Plotly
fig = px.bar(
    avg_order_value_by_payment,
    x='average_order_value',
    y='payment_method',
    orientation='h',
    title='Average Order Value by Payment Method',
    labels={'average_order_value': 'Average Order Value', 'payment_method': 'Payment Method'}
)
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

This cell compares total shipping costs across different product categories. It groups the DataFrame by 'category', sums the 'shipping_cost' for each, and then displays the results. A horizontal bar chart is generated using Plotly Express to visually represent the total shipping costs per category.

```python
# Calculate total shipping costs for each category
category_shipping_costs = df.groupby('category')['shipping_cost'].sum().reset_index()
category_shipping_costs.columns = ['category', 'total_shipping_cost']

# Display the shipping costs by category
display(category_shipping_costs)

# Plot Shipping Costs by Category using Plotly
fig = px.bar(
    category_shipping_costs,
    x='total_shipping_cost',
    y='category',
    orientation='h',
    title='Total Shipping Costs by Category',
    labels={'total_shipping_cost': 'Total Shipping Cost', 'category': 'Category'}
)
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```
